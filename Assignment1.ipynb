# Overview of NYC311 dataset

clean data for general analysis

```python
Data1 = df_pandas[['Created Date','Closed Date','Agency','Agency Name','Complaint Type','Descriptor','Borough','Incident Zip']]
```

clean columns and data types

```python
Data1.rename(columns = {'Incident Zip':'Zipcode'}, inplace = True) 
Data1['Zipcode'] = Data1['Zipcode'].astype(str).str[0:5]
```

```python
Data1['Created Date'] = pd.to_datetime(Data1['Created Date'],format ='%m/%d/%Y %I:%M:%S %p')
Data1['Closed Date'] = pd.to_datetime(Data1['Closed Date'],format ='%m/%d/%Y %I:%M:%S %p')
```

Monthly situation check

```python
Mton['MONTH'] = pd.to_datetime(Mton['MONTH'],format ='%Y / %m')
Mton_2019=Mton[Mton['MONTH'].dt.year==2019]
Mton_2019
```

```python
Tonnage_Borough = Mton_2019[['MONTH','BOROUGH']]
```

```python
Tonnage = Mton_2019[['REFUSETONSCOLLECTED','PAPERTONSCOLLECTED','MGPTONSCOLLECTED','RESORGANICSTONS','SCHOOLORGANICTONS','LEAVESORGANICTONS','XMASTREETONS']]
Tonnage
```

```python
Tonnage_sum = Tonnage.sum(axis=1)
Tonnage_sum
```

```python
Borough_tonnage_sum = pd.concat([Tonnage_Borough, Tonnage_sum], axis=1)
Borough_tonnage_sum
```

```python
Manhattan_tonnage = Borough_tonnage_sum[(Borough_tonnage_sum['BOROUGH']=='Manhattan')].groupby('MONTH').sum('0')
```

```python
Brooklyn_tonnage = Borough_tonnage_sum[(Borough_tonnage_sum['BOROUGH']=='Brooklyn')].groupby('MONTH').sum('0')
```

```python
Queens_tonnage = Borough_tonnage_sum[(Borough_tonnage_sum['BOROUGH']=='Queens')].groupby('MONTH').sum('0')
```

```python
StatenIsland_tonnage = Borough_tonnage_sum[(Borough_tonnage_sum['BOROUGH']=='Staten Island')].groupby('MONTH').sum('0')
```

```python
Bronx_tonnage = Borough_tonnage_sum[(Borough_tonnage_sum['BOROUGH']=='Bronx')].groupby('MONTH').sum('0')
```

```python
Mton_cleaned = pd.concat([Manhattan_tonnage, Brooklyn_tonnage,Queens_tonnage,StatenIsland_tonnage,Bronx_tonnage], axis=1)
Mton_cleaned.columns=['Manhattan','Brooklyn','Queens','Staten Island','Bronx']
Mton_cleaned['Total'] = Mton_cleaned.sum(axis=1)
Mton_cleaned.sort_values(by='Total',ascending = True)
```
